# Dataset 1

```python
import pandas as pd
# import data 
dataframe_train1 = pd.read_csv (r'/home/yupaporn/codes/USAI/Traindf_fold1.csv')
dataframe_val1 = pd.read_csv (r'/home/yupaporn/codes/USAI/Validationdf_fold1.csv')
# dataframe_test1 = pd.read_csv (r'/home/yupaporn/codes/USAI/Testdf.csv')
dff1 = pd.concat([dataframe_val1,dataframe_train1])
print(len(dff1))
```

```python
 dff1[["Class","Case"]].groupby(['Class']).count()
```

```python
 dff1[["Class","Views","Case"]].groupby(['Views','Class']).count()
```

```python
 dff1[["Views","Case"]].groupby(['Views']).count()
```
